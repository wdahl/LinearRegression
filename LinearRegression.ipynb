```python
#William Dahl
#ICSI 436
#HW2
#Febuary 18th, 2019

import numpy
import math
import matplotlib.pyplot as plt

#Mave a Vandermonde matrix
def make_matrix_X(x, degree):
    #intializes the matrix with all zeros
    X = numpy.matrix(numpy.zeros((len(x), degree+1)))
    #loops through the values in x
    for i in range(len(x)):
        #goes the degree specified 
        for j in range(degree+1):
            #calculates the matrix
            X[i,j] = math.pow(x[i], j)
                
    return X #the matrix

#The polynominal function f
def f(a, x):
    y = 0
    #caluulates the estemated value of y
    for i in range(len(a)):
        y += a[i,0] * math.pow(x,i)
        
    return y #estimate 

#graphs the line bassed on a and x
def graph(a, x):
    y = []
    for i in range(len(x)):
        y.append(f(a, x[i]))
        
    plt.plot(x,y,'b')
    plt.show()

#calulates the error of the estimate 
def error(x,y,a):
    error = 0
    for i in range(len(x)):
        error += math.pow(y[i] - f(a,x[i]), 2)
        
    return error/len(x)
    
x = []
y = []
#reads the data file
for line in open('hw2.dat'):
    point = line.split()
    x.append(float(point[0]))
    y.append(float(point[1]))

#makes the y values into a numpy matrix for easy calulations
y_vector = numpy.matrix(y)

#makes the X matrix with degree 1
X = make_matrix_X(x, 1)
#Calculates the values of a
a = (X.getT()*X).getI() * X.getT() * y_vector.getT()
print('d = ', 1)
print("a = ", a)
print('error = ',error(x,y,a))
plt.plot(x,y, 'ro')
graph(a,x)


X = make_matrix_X(x, 3) #degree 3
a = (X.getT()*X).getI() * X.getT() * y_vector.getT()
print('d = ', 3)
print("a = ", a)
print('error = ',error(x,y,a))
plt.plot(x,y, 'ro')
graph(a,x)

X = make_matrix_X(x, 5)#degree 5
a = (X.getT()*X).getI() * X.getT() * y_vector.getT()
print('d = ', 5)
print("a = ", a)
print('error = ',error(x,y,a))
plt.plot(x,y, 'ro')
graph(a,x)

X = make_matrix_X(x, 7)#degree 7
a = (X.getT()*X).getI() * X.getT() * y_vector.getT()
print('d = ', 7)
print("a = ", a)
print('error = ',error(x,y,a))
plt.plot(x,y, 'ro')
graph(a,x)
```
